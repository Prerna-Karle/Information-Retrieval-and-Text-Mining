```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re

# Input two documents
doc1 = input("Enter text for Document 1:\n")
doc2 = input("\nEnter text for Document 2:\n")

# Function to remove stop words
def remove_stopwords(text):
  # Convert to Lowercase
  text = text.lower()

  # Remove punctuation and numbers
  words = re.findall(r'\b[a-z]+\b', text)

  # Remove stop words
  filtered_words = [word for word in words if word not in ENGLISH_STOP_WORDS]

  return " ".join(filtered_words)

# Process both documents
processed_doc1 = remove_stopwords(doc1)
processed_doc2 = remove_stopwords(doc2)

# Output results
print("\n--- After Stop Word Removal ---")
print("\nDocument 1:")
print(processed_doc1)

print("\nDocument 2:")
print(processed_doc2)
```
