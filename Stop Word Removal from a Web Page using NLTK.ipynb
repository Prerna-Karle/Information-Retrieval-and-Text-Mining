```python
# Step 1 : Import Libraries 
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

```python
# Step 2 : Download Required NLTK Resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# Step 3 : Fetch Web Page Content 
# URL to scrape
url = "https://oxylabs.io/"

# Send request
response = requests.get(url)

# Check status
print("Status Code:", response.status_code)
```

```python
# Step 4 : Parse HTML and extract text 
# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Extract visible text
text = soup.get_text()

# Display small sample
print(text[:500])
```

```python
# Step 5 : Save Scraped Text 
with open("scraped_text.txt", "w", encoding="utf-8") as file:
    file.write(text)

print("Scraped text saved successfully.")
```

```python
# Step 6 : Text Cleaning 
# Convert to lowercase
text = text.lower()

# Remove special characters and numbers
text = re.sub(r'[^a-z\s]', '', text)
```

```python
# Step 7 : Tokenization using NLTK
# Tokenize text into words
tokens = word_tokenize(text)

print("Total words before stop word removal:", len(tokens))
```

```python
# Step 8 : Stop Word Removal using NLTK
# Load English stop words
stop_words = set(stopwords.words('english'))

# Remove stop words
filtered_words = [word for word in tokens if word not in stop_words]

print("Total words after stop word removal:", len(filtered_words))
```

```python
# Step 9 : Join Words Back into Text
processed_text = " ".join(filtered_words)

print("\nSample Output After Stop Word Removal:\n")
print(processed_text[:500])
```
