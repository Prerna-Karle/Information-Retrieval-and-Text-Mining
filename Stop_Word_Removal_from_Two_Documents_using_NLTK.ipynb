```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Load stop words
stop_words = set(stopwords.words('english'))

# Input two documents
doc1 = input("Enter text for Document 1:\n")
doc2 = input("\nEnter text for Document 2:\n")

def remove_stopwords(text):
    text = text.lower()
    tokens = word_tokenize(text)
    filtered = [word for word in tokens if word.isalpha() and word not in stop_words]
    return " ".join(filtered)

# Process documents
processed_doc1 = remove_stopwords(doc1)
processed_doc2 = remove_stopwords(doc2)

# Output
print("\n--- After Stop Word Removal ---")
print("\nDocument 1:")
print(processed_doc1)

print("\nDocument 2:")
print(processed_doc2)
```
