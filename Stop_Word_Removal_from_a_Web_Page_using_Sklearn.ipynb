```python
import requests
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re
```

```python
# Fetch Web Page Content
url = "https://oxylabs.io/"
response = requests.get(url)

print("Status Code:", response.status_code)
```

```python
# Step 2 : Parse HTML using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

text = soup.get_text()

print(text[:500])
```

```python
# Step 3 : Save scraped text to a file
with open("scraped_text.txt", "w", encoding="utf-8") as file:
    file.write(text)

print("Web page text saved successfully.")
```

```python
# Step 4 : Text Proprocessing

# Convert to lowercase
text = text.lower()

# Remove special characters and numbers
text = re.sub(r'[^a-z\s]', '', text)

# Tokenize
words = text.split()

print("Total words before stop word removal:", len(words))
```

```python
# Step 5 : Stop Word Removal using sklearn
filtered_words = [word for word in words if word not in ENGLISH_STOP_WORDS]

print("Total words after stop word removal:", len(filtered_words))
```

```python
# Step 6 : Final Processed Text

# Join words back into text
processed_text = " ".join(filtered_words)

# Display sample output
print("\nSample Output After Stop Word Removal:\n")
print(processed_text[:500])
```

```python
# Step 7 : Save Final Output

# Save processed text
with open("processed_text_no_stopwords.txt", "w", encoding="utf-8") as file:
    file.write(processed_text)

print("Processed text saved successfully.")
```
